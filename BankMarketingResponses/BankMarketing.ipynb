<h3><b>3.2 IQR and data filtering based on IQR</b><h3>

### Function to calculate the Inter quartile range for quantile values outside 25 to 75 range. And to do the outlier capping for lower level with min value and for upper level with 'q3=1.5*iqr' value.

````python
def remove_outlier(df_in, col_name):
    Q1=df_in[col_name].quantile(0.25)
    Q3=df_in[col_name].quantile(0.75)
    # Calculating IQR as difference of Quantile 3 and quantile 1
    IQR=Q3-Q1
    min_col=df_in.min()
    max_col=df_in.max()
    lower_bound=df_in[col_name].min()
    upper_bound=Q3 + 1.5*IQR
    print('Column',col_name,'IQR lower bound and upper bound are', lower_bound, 'and', upper_bound, 'respectively')
    df_out=df_in[(df_in[col_name]>lower_bound) & (df_in[col_name]<upper_bound)]
    return df_out
````

### Using the above created function , remove the outlier from 'age' variables:

````python
def age_outlier(bank):
    # your code to remove the outlier from age
    col_name='age'
    df=remove_outlier(bank,col_name)
    return df
    
bank = age_outlier(bank) 
````

### Using the above created function , remove the outlier from 'campaign' variables:

````python
def campaign_outlier(bank):
    col_name='campaign'
    df=remove_outlier(bank,col_name)
    return df

    
bank = campaign_outlier(bank) 
````

### Using the above created function , remove the outlier from 'duration' variables:

<h2><b>4.1 Building Logistic Regression Model</b></h2>

### Instantitating Logistic Regression model using scikit-learn

````python
from sklearn.linear_model import LogisticRegression
def model_initialization():
    # initiate the logistic regression model to new variable logreg
    logreg=LogisticRegression()
    return logreg
logreg = model_initialization()
````

### Fitting the logistic model on X_train and y_train

````python
def model_fit():
    # fit the X_train and y_train
    logreg.fit(X_train,y_train)
model_fit()    
````

### Using the model for prediction

````python
y_pred_train = pd.DataFrame()
def logistic_prediction_train():
    # create the model prediction on X_train data using the above created dataframe
    y_pred_train=logreg.predict(X_train)
    return y_pred_train
y_pred_train = logistic_prediction_train()
````

````python
y_pred_test = pd.DataFrame()
def logistic_prediction_test():
    # create the model prediction on X_test data using the above created dataframe
    y_pred_test=logreg.predict(X_test)
    return y_pred_test
y_pred_test = logistic_prediction_test()
````

<h3><b>4.2 Model evaluation</b></h3>

### Model evaluation using accuracy classification score

````python
from sklearn.metrics import accuracy_score
def logistic_accuracy_score():
    #Calculate and print the accuracy score
    print("The accuracy of test_prediction w.r.t actual_test is :",accuracy_score(y_test,y_pred_test))
    
logistic_accuracy_score()
````

### Model evaluation using Confusion matrix

````python
from sklearn.metrics import confusion_matrix
confusion_matrix = pd.DataFrame(confusion_matrix(y_test, y_pred_test))

def model_evaluation():
    # provide confusion matrix index name as['Actual No_Deposit','Actual Deposit']
    # provide confusion matrix index name as['Predicted No_Deposit','Predicted Deposit']
    confusion_matrix.index=['Actual No_Deposit','Actual Deposit']
    confusion_matrix.columns=['Predicted No_Deposit', 'Predicted Deposit']
    print(confusion_matrix)
model_evaluation()
````
